# SQL Alchemy

```python
# Import SQL Alchemy for connecting to DBs
from sqlalchemy import create_engine, inspect

# Import and establish Base for which classes will be constructed
from sqlalchemy.orm import declarative_base
Base = declarative_base()

# Import modules to declare columns and data types
from sqlalchemy import Column, Integer, String, Float, ForeignKey, func
```

```python
# Create DB Engine
engine = create_engine("sqlite:///flights.db")
```

```python
# Create a Session Object to Connect to DB
from sqlalchemy.orm import Session
session = Session(bind=engine)
session
```

```python
# Reset everything: drop tables, clear mappings, and close session
Base.metadata.drop_all(engine)
Base.metadata.clear()

try:
    session.rollback()
except:
    pass

session.close()
```

```python
session

# Create an Airline Class
class Airline(Base):
    __tablename__ = 'airlines'
    IATA_code = Column(String(3), primary_key=True)
    Airline = Column(String(30))

# Create an Airport Class
class Airport(Base):
    __tablename__ = 'airports'
    iata_code = Column(String(3), primary_key=True)
    name = Column(String(30))

# Create an Airline Rating Class
class AirlineRating(Base):
    __tablename__ = 'airline_rating'
    id = Column(Integer, primary_key=True)
    AirlineName = Column(String(30))
    TravelType = Column(String(30))
    CabinType = Column(String(30))
    Month = Column(Integer)
    DayOfWeek = Column(String(30))
    Avg_OverallScore = Column(Float)
    Avg_SeatComfortRating = Column(Float)
    Avg_EntertainmentRating = Column(Float)
    Avg_FoodRating = Column(Float)
    Avg_GroundServiceRating = Column(Float)
    Avg_ServiceRating = Column(Float)
    Avg_ValueRating = Column(Float)	
    Avg_WifiRating = Column(Float)

# Create an Flight Data Class
class FlightData(Base):
    __tablename__ = 'flight_data'
    id = Column(Integer, primary_key=True)
    FL_DATE = Column(String(30))
    OP_CARRIER = Column(String(2), ForeignKey('airlines.IATA_code'))
    OP_CARRIER_FL_NUM = Column(Integer)
    ORIGIN = Column(String(3), ForeignKey('airports.iata_code'))
    DEST = Column(String(3), ForeignKey('airports.iata_code'))
    CRS_DEP_TIME = Column(Integer)
    DEP_TIME = Column(Float)
    DEP_DELAY = Column(Float)
    CRS_ARR_TIME = Column(Integer)
    ARR_TIME = Column(Float)
    CANCELLED = Column(Float)
    AIR_TIME = Column(Float)
    DISTANCE = Column(Float)
    CARRIER_DELAY = Column(Float)
    WEATHER_DELAY = Column(Float)
    NAS_DELAY = Column(Float)
    SECURITY_DELAY = Column(Float)
    LATE_AIRCRAFT_DELAY = Column(Float)
    Month = Column(Integer)
    DayOfWeek = Column(String(30))

# Create tables
Base.metadata.create_all(engine)
```

```python
query = """
    SELECT 
        r.AirlineName,
        a.IATA_code AS CarrierCode,
        f.Month,
        f.DayOfWeek,
        r.TravelType,
        r.CabinType,
        
        -- Ratings
        r.Avg_OverallScore,
        r.Avg_SeatComfortRating,
        r.Avg_EntertainmentRating,
        r.Avg_FoodRating,
        r.Avg_GroundServiceRating,
        r.Avg_ServiceRating,
        r.Avg_ValueRating,
        r.Avg_WifiRating,
        
        -- Reliability
        AVG(f.DEP_DELAY) AS Avg_DepartureDelay,
        AVG(f.CANCELLED) AS CancelRate,
        AVG(f.AIR_TIME) AS Avg_AirTime,
        AVG(f.DISTANCE) AS Avg_Distance,
        
        -- Airport
        f.ORIGIN AS OriginAirport

    FROM airline_rating r
    JOIN airlines a ON r.AirlineName = a.Airline
    JOIN flight_data f ON a.IATA_code = f.OP_CARRIER

    GROUP BY 
        r.AirlineName, a.IATA_code, f.Month, f.DayOfWeek, 
        r.TravelType, r.CabinType, f.ORIGIN,
        r.Avg_OverallScore, r.Avg_SeatComfortRating,
        r.Avg_EntertainmentRating, r.Avg_FoodRating,
        r.Avg_GroundServiceRating, r.Avg_ServiceRating,
        r.Avg_ValueRating, r.Avg_WifiRating
"""
```

```python
analysis_df = pd.read_sql(query, con=engine)
analysis_df
```
